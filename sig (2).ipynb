### Removing stopwords and stemming the remaining words

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
stem = []
for i in reviews:
    i = i.lower()
    alpha = re.sub('[^a-zA-Z]', ' ', i)
    alpha_list = alpha.split()
    temp = []
    sw = set(stopwords.words('english'))
    for j in alpha_list:
        if(j not in sw):
            temp.append(ps.stem(j))
    string = " ".join(temp)
    stem.append(string)
```
